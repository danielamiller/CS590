```python
def is_prime(num):
    for n in range(2,num):
        if num % n == 0:
            print(num, 'is not prime')
            break
    else:
        print(num, 'is prime')
```

```python
is_prime(7)
```

```python
is_prime(16)
```

```python
is_prime(19348)
```

```python
is_prime(1529383)
```

```python
import math

def is_prime2(num):
    for n in range(2,num):
        if num % n == 0 and num >2:
            return False
    for i in range(3, int (math.sqrt(num))+1,2):
        if num % i == 0:
            return True
    return True
```

```python
is_prime(13487)
```

```python
is_prime2(13487)
```

```python
#global statements
name = 'This is a global name'

def greet():
    #enclosing function
    name = 'Sammy'
    def hello():
        print('Hello', name)
        
    hello()
    
greet()
```

```python
greet()
#this still runs the hello function when running greet since nested within
```

```python
print(name)
```

```python
#can change globals within a function

x=50

def func():
    global x
    #makes sure that function available after it is run
    print("Because of the global declaration, x is", x)
    x=2
    print('Ran func(), changed global to', x)
```

```python
print('before calling func, x is',x)
```

```python
func()
```
