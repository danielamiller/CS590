```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
#merging dataframes
#left = 'left'
#right = 'right'
#inner = 'inner'
#outer = 'outer'
# Merge Left and Right on the Key and only keep the rows where key values match
pd.merge(left, right, how='inner', on='key')
```

```python
#more complexity to merge; 2 keys
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
# Using two keys
pd.merge(left, right, on=['key1','key2'])
#default is inner join
#inner join merges only rows with matching keys between both datasets
#outer join merges all lines, regardless if shared
#creates a cartesian product -> includes all possibilies for key pair matches
#note K1 K0 has two outputs with different values
```

```python
# Use an outer merge
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
####JOINING
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
left.join(right)
#note K3 line not present because joining based on left dataframe only
```

```python
#join merge based on left and right only; merge has more flexibility
#same thing different way
df1 = left
df2 = right
df1.join(df2)
```
